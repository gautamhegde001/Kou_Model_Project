```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

input_dir = Path.cwd().parent / 'datafiles'

plt.rcParams['figure.facecolor']='white'
plt.figure(figsize=(10,6),dpi=200)


lam_values = [0.0,1.5,6.0]
colors = ['r','g','b']
for i,lam in enumerate(lam_values) :
    input_filename = f'simulated_PnL_delta_hedging_lam={lam}.npy'
    input_path = input_dir/input_filename

    Data = np.load(input_path)
    print(np.mean(Data))
    print(len(Data))
    plt.hist(Data,bins=40,density=True,histtype='step',linewidth=2,label=f'$\lambda={lam}$',color=colors[i])
    plt.axvline(np.mean(Data),linestyle='dashed',color = colors[i],linewidth = 1)
plt.legend(fontsize=20)
plt.ylabel("Probability Density",fontsize=24)
plt.xlabel("Portfolio Value ($)",fontsize=24)
plt.tick_params(labelsize=16)
plt.title(f'Histogram of P&L for $S_0=100 \$, K = 80 \$ $',fontsize=24)
plt.xlim(-25,200)
plt.show()
```
