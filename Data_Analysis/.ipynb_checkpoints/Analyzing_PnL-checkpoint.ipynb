```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

input_dir = Path.cwd().parent / 'datafiles'

plt.rcParams['figure.facecolor']='white'
plt.figure(figsize=(10,6))


lam_values = [0.0,1.5,6.0]
for lam in lam_values :
    input_filename = f'simulated_PnL_delta_hedging_lam={lam}.npy'
    input_path = input_dir/input_filename

    Data = np.load(input_path)
    print(np.mean(Data))
    plt.hist(Data,bins=50)
plt.ylabel("No. of simulations",fontsize=24)
plt.xlabel("Portfolio Value ($)",fontsize=24)
plt.tick_params(labelsize=16)
plt.title(f'Histogram of P and L for $\lambda={lam}$',fontsize=24)
plt.xlim(-25,200)
plt.show()
```
