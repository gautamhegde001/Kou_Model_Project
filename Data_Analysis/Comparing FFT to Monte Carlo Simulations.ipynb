```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.rcParams['text.usetex'] = True

input_dir = Path.cwd().parent / 'datafiles'

T_values = [0.5,1.0,2.0]
colors = ['b','r','g']
fig,ax = plt.subplots(figsize=(10,6),dpi=200)
for i,T in enumerate(T_values) :

    input_filename_fft = f'strikes_vs_prices_kou_fft_T={T}.npy'
    input_filename_monte_carlo = f'strikes_vs_prices_kou_monte_carlo_T={T}.npy'

    input_path_fft = input_dir/input_filename_fft
    input_path_monte_carlo = input_dir/input_filename_monte_carlo


    strikes_fft,prices_fft = np.load(input_path_fft)
    strikes_mc,prices_mc = np.load(input_path_monte_carlo)


    ax.plot(strikes_fft,prices_fft,label=f'Via FFT : T={T}',linewidth=2,color = colors[i])

    ax.plot(strikes_mc,prices_mc,'o',label=f'Via Monte Carlo : T={T}',markersize=5,color = colors[i])

    ax.set_xlabel(r"Strike Prices $K$ (\$)",fontsize=24)

    ax.set_ylabel(r"Option Prices(\$)",fontsize=24)
ax.tick_params(axis='both',labelsize=16)

ax.set_title(r"Option prices for $S_0=100.0 \$ $",fontsize=24)


ax.legend(fontsize=16)
fig.set_facecolor('White')
plt.show()
```
